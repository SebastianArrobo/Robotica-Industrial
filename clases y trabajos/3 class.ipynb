```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x=np.arange(-10,10,.1)
y = np.sin(x)
z = np.cos(x)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.plot(x,y,z,c='red')
ax.view_init(20,70)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-10,10,.1)
y = np.arange(-10,10,.1)

xv,yv = np.meshgrid(x,y)
zv = np.sin(xv)

ax.plot_wireframe(xv,yv,zv)
```
